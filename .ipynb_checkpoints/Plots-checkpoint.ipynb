```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
raw_data = pd.read_csv("titanic\\train.csv")
raw_data
```

```python
sns.set(rc={'figure.figsize': (12,10)})
```

```python
ax = sns.lineplot(x = 'Age', y = 'Fare', data = raw_data, hue = 'Embarked', ci = False, markers = True)
```

```python
ax = sns.lineplot(x = 'Age', y = 'Fare', data = raw_data, hue = 'Sex', ci = False)
```

```python
ax = sns.barplot(x= 'Survived', y = 'Sex', data = raw_data, orient = 'h')
print("Aggregate over sum")
print(raw_data[["Survived", "Sex"]].groupby('Sex').agg({"Survived": 'sum'}))
print("Aggregate over mean")
print(raw_data[["Survived", "Sex"]].groupby('Sex').agg({"Survived": 'mean'}))
```

```python
ax = sns.barplot(x= 'Sex', y = 'Survived', data = raw_data, hue = 'Pclass')
```

```python
x = raw_data['Fare'].values
mean = raw_data['Fare'].mean()
sns.distplot(x, color = 'blue');
plt.axvline(mean, 0, 1, color = "red")
print("The line in the red in the mean line")
```

```python
x = raw_data['Age'].values
mean = raw_data['Age'].mean()
sns.distplot(x, color = 'red');
plt.axvline(mean, 0, 1, color = "black")
print("The line in the red in the mean line")
```

```python
ax = sns.boxplot(x = 'Pclass', y = 'Age', data = raw_data)
#ax = sns.swarmplot(x = 'Pclass', y = 'Fare', data = raw_data, color = 'red')
```

```python
ax = sns.boxplot(x = 'Survived', y = 'Age', data = raw_data)
```

```python
ax = sns.scatterplot(x = 'Age', y = 'Fare', data = raw_data, color = 'green')
```

```python
ax = sns.scatterplot(x = 'Age', y = 'Fare', data = raw_data, hue = 'Embarked', style='Sex')
```

```python
ax = sns.lmplot(x = 'Age',y = 'Fare',data = raw_data, col = 'Pclass', ci = None, line_kws = {'color':'black'})
```

```python
ax = sns.lmplot(x = 'Age',y = 'Fare',data = raw_data, col = 'Embarked', ci = False, line_kws = {'color':'black'})
```

```python
ax = sns.lmplot(x = 'Age',y = 'Fare',data = raw_data, col = 'Sex', ci = False, line_kws = {'color':'red'})
```

```python
fig, axis = plt.subplots(2,2, figsize = (12,7))

a = raw_data['Fare'].values
b = raw_data['Age'].values
c = raw_data['Pclass'].values
d = raw_data['Survived'].values
sns.distplot(a, color = 'blue', ax = axis[0,0]);
sns.distplot(b, color = 'blue', ax = axis[0,1]);
sns.barplot(x= 'Survived', y = 'Sex', data = raw_data, ax = axis[1,0])
sns.barplot(x= 'Survived', y = 'Embarked', data = raw_data, ax = axis[1,1])
```

```python
g= sns.pairplot(raw_data, plot_kws={'color': 'green'})
```

```python
ax = sns.jointplot('Age', 'Fare', data = raw_data, height = 10, kind = "reg")
```

```python
fig , axis = plt.subplots(2,1)
cd1 = raw_data[['Age', 'Fare', 'Pclass']].corr()
cd2 = raw_data[['Age', 'Fare', 'Survived']].corr()
sns.heatmap(cd1, ax = axis[0])
sns.heatmap(cd2, ax = axis[1])
```
